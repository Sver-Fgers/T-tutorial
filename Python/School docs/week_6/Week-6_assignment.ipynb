```python
# 2a 
# using regex to find patterns in code

# import regex module
import re

# function to find the first occurrence
def find_sequence(sequence, genome, name):
    match = re.search(sequence, genome)
    if match:
        print(f"In {name}, '{sequence}' first appears at position {match.start()}")
    else:
        print(f"In {name}, '{sequence}' was not found")

# Perform regex search
find_sequence('ATGCATGC', r_genome_1, 'r_genome_1')
find_sequence('ATGCATGC', r_genome_2, 'r_genome_2')
find_sequence('ATGCATGC', r_genome_3, 'r_genome_3')
```

```python
# 2b

patterns = ['AAAA', 'TTTT', 'CCCC', 'GGGG']

# Randomly select an index from the list
pattern_select = random.randint(0, len(patterns) - 1)

# Select the pattern at the random index
selected_pattern = patterns[pattern_select]


def find_first_pattern(patterns, genome, name):
    for pattern in patterns:
        match = re.search(pattern, genome)
        if match:
            print(f"In {name}, '{selected_pattern}' first appears at position {match.start()}")
            return
    print(f"In {name}, none of the patterns were found")

find_first_pattern(patterns, r_genome_1, 'r_genome_1')
find_first_pattern(patterns, r_genome_2, 'r_genome_2')
find_first_pattern(patterns, r_genome_3, 'r_genome_3')
```

```python
# 3a

def find_identical_sequences(genome, name):
    # Find matching sequences of 8 or more identical bases (A, T, C, G)
    matches = re.finditer(r"(A{5,}|T{5,}|C{5,}|G{5,})", genome)
    found = False
    
    # Loop to check if any matches exist
    for match in matches:
        if not found:
            print(f"In {name}:")  # Only print this once for each genome
            found = True
        start = match.start()
        length = len(match.group())
        print(f"   '{match.group()}' at position {start} (length: {length})")
    
    # If no matches found, print this line once for each genome
    if not found:
        print(f"In {name}, no sequences of 8 or more identical characters were found.")

# Call the function for each genome
find_identical_sequences(r_genome_1, 'r_genome_1')
find_identical_sequences(r_genome_2, 'r_genome_2')
find_identical_sequences(r_genome_3, 'r_genome_3')
```

```python
# 3b

def find_identical_sequences(genome, name):
    # Find matching sequences of 8 or more identical bases (A, T, C, G)
    matches = re.finditer(r"(A{8,}|T{8,}|C{8,}|G{8,})", genome)
    found = False
    
    # Loop to check if any matches exist
    for match in matches:
        if not found:
            print(f"In {name}:")  # Only print this once for each genome
            found = True
        start = match.start()
        length = len(match.group())
        print(f"   '{match.group()}' at position {start} (length: {length})")
    
    # If no matches found, print this line once for each genome
    if not found:
        print(f"In {name}, no sequences of 8 or more identical characters were found.")

# Call the function for each genome
find_identical_sequences(r_genome_1, 'r_genome_1')
find_identical_sequences(r_genome_2, 'r_genome_2')
find_identical_sequences(r_genome_3, 'r_genome_3')


```

```python
# 3b
def find_longest_no_c(genome, name):
    matches = re.finditer(r"[^C]+", genome)
    longest = max(matches, key=lambda m: len(m.group()), default=None)
    if longest:
        start = longest.start()
        length = len(longest.group())
        sequence = longest.group()[:20] + "..." if length > 20 else longest.group()
        print(f"In {name}: Longest substring without 'C' starts at position {start} (length: {length}) - Sequence: '{sequence}'")
    else:
        print(f"In {name}, no sequences without 'C' were found.")

find_longest_no_c(r_genome_1, 'r_genome_1')
find_longest_no_c(r_genome_2, 'r_genome_2')
find_longest_no_c(r_genome_3, 'r_genome_3')
```

```python
# 4a

def find_start_positions(genome):
    # Find the start positions of "ATG" in the genome
    start_positions = [m.start() for m in re.finditer("ATG", genome)]
    
    # Check if no start codon is found
    if not start_positions:
        print("No 'ATG' start codons found.")
    else:
        print(f"ATG start codon positions:")
        print(f"  {start_positions}")
    
    return start_positions

# testing with r_genome_1
start_position_list = find_start_positions(r_genome_1)
```

```python
# 4b

def find_stop_positions(genome):
    stop_positions = [m.start() for m in re.finditer("TAA|TAG|TGA", genome)]

    # Check if no start codon is found
    if not stop_positions:
        print("No stop codons found.")
    else:
        print(f"Stop codon positions: {stop_positions}")
    return stop_positions

end_position_list = find_stop_positions(r_genome_1)
```

```python
#4c

def find_orfs(start_positions, stop_positions):
    for start in start_positions:
        subsequent_stops = [stop for stop in stop_positions if stop > start]
        if subsequent_stops:
            closest_stop = min(subsequent_stops)
            print(f"ORF from {start} to {closest_stop}")
        else:
            print(f"No stop codon found for start codon at {start}")

find_orfs(start_position_list, end_position_list)
```

```python
# 4d

def find_orfs_with_regex(genome):
    orfs = re.finditer(r"ATG(?:[ATGC]{3})*?(?:TAA|TAG|TGA)", genome)
    for orf in orfs:
        start = orf.start()
        end = orf.end()
        sequence = orf.group()[:20] + "..." if len(orf.group()) > 20 else orf.group()
        print(f"ORF from {start} to {end} - Sequence: '{sequence}'")

find_orfs_with_regex(r_genome_1)
```
