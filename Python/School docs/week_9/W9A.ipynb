```python
# 1
def load_wordset(filename):
    """
    Reads a file containing words and returns a set of unique words.
    
    args:
    filename (str): Path to the file containing the words.
    
    Returns:
    set: A set containing all the unique words from the file.
    """
    try:
        with open(filename, 'r') as file:
            words = file.read().splitlines()
        return set(words)
    except FileNotFoundError:
        print(f"Error: The file '{filename}' does not exist.")
        return set()

# usage
filepath = r"C:\Users\Victor_Nxx\Documents\.Programming\.Tutoring\T\3000words.txt"
wordset = load_wordset(filepath)
print(wordset)
```

```python
# 2
import matplotlib.pyplot as plt

def create_word_length_histogram(wordset):
    """
    Creates a histogram of word lengths from a given wordset.
    
    Args:
    wordset (set): A set of words.
    """
    word_lengths = [len(word) for word in wordset]
    plt.hist(word_lengths, bins=range(1, max(word_lengths) + 2), color='skyblue', edgecolor='black', alpha=0.7)
    
    plt.title("Histogram of Word Lengths")
    plt.xlabel("Word Length")
    plt.ylabel("Frequency")
    plt.xticks(range(1, max(word_lengths) + 1))
    plt.show()

# Usage
create_word_length_histogram(wordset)
```

```python
# 3
def get_words_of_len(wordset, wordlen):
    """
    Returns a list of words from the wordset with the specified length.

    Args:
    wordset (set): A set of words.
    wordlen (int): The desired word length.

    Returns:
    list: A list of words with the specified length.
    """
    return [word for word in wordset if len(word) == wordlen]

# Usage
wordlist = get_words_of_len(wordset, 2)
print(wordlist)
```

```python
# 4
import pandas as pd
import matplotlib.pyplot as plt

# Load data from the file
file_path = r'C:\Users\Victor_Nxx\Documents\.Programming\.Tutoring\T\ktup_approximation.txt'

# Reading the data into a DataFrame
data = pd.read_csv(file_path, sep='\t')

# Renaming columns for clarity
data.columns = ['Global Pairwise Identity (%)', 'Shared k-tuples (%)']

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Global Pairwise Identity (%)'], 
            data['Shared k-tuples (%)'], 
            c='blue', label='k-tuples (k=2)', alpha=0.7, edgecolors='black')

# Adding titles and labels
plt.title('Scatter Plot of Shared k-tuples vs. Global Pairwise Identity', fontsize=14)
plt.xlabel('Global Pairwise Identity (%)', fontsize=12)
plt.ylabel('Shared k-tuples (%)', fontsize=12)

# Adding grid, legend, and showing the plot
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.show()
```

```python
# 5 & 6 
import numpy as np
import matplotlib.pyplot as plt

with open(r'C:\Users\Victor_Nxx\Documents\.Programming\.Tutoring\T\bacteria_abundances_W7.txt', 'r') as file:
    lines = file.readlines()
    species = lines[0].strip().split(", ")  # First line contains species names
    data = np.loadtxt(lines[1:], delimiter=",")  # Remaining lines are numeric data


# Separate data into Crohn's and non-Crohn's
crohns_data = data[:5]
non_crohns_data = data[5:]

# Calculate mean and standard deviation for each group
crohns_mean = crohns_data.mean(axis=0)
crohns_std = crohns_data.std(axis=0)

non_crohns_mean = non_crohns_data.mean(axis=0)
non_crohns_std = non_crohns_data.std(axis=0)

# Plotting
x = np.arange(len(species))  # Positions for bars
width = 0.4  # Width of bars

plt.figure(figsize=(14, 7))

# Bars for Crohn's patients
plt.bar(x - width/2, crohns_mean, width, yerr=crohns_std, capsize=5, label="Crohn's Patients", color='skyblue')

# Bars for non-Crohn's patients
plt.bar(x + width/2, non_crohns_mean, width, yerr=non_crohns_std, capsize=5, label="Non-Crohn's Patients", color='salmon')

# Add labels, title, and legend
plt.xticks(x, species, rotation=45, ha='right', fontsize=10)
plt.ylabel("Average Abundance", fontsize=12)
plt.title("Average Abundance of Bacterial Species/Genera", fontsize=14)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show plot
plt.show()
```
