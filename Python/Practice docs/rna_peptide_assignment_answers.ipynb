```python
#1

import pandas as pd 
# read the csv file
df = pd.read_csv("/Users/bosslistic10/Documents/t-programming/rna_weight.csv", sep = ",", index_col = 0)
# display to see content of file
display(df)

s = "AUGGCUAACUGG"
# initialize variable to add 
fw = 0
# loop through s, locate mw, the add into fw
for i in s:
    fw += df.loc[i, "MolecularWeight"]
# substract 18 from the total no of bonds  
fw -= 18 * (len(s) - 1)
# print the result
print(f"Molecular weight of RNA: {fw:.1f} g/mol")
    
```
